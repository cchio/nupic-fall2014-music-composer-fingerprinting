```python
from music21 import *
import numpy as np
import pandas as pd
import music21
import csv
```

```python
score = converter.parse('/Users/zichong/Desktop/taylorswift/taylor_swift-love_story.mid')
```

```python
score.flat.show('text')[:50]
```

```python
partStream = score.parts
print partStream.getContextByClass('TimeSignature')
print partStream.getContextByClass('KeySignature')

with open('/Users/zichong/Desktop/nupic.examples/sine-prediction/taylor_loves_clarence.csv', 'wb') as csvfile:
    mywriter = csv.writer(csvfile, delimiter=',')
    mywriter.writerow(['twelvth_beat', 'pitch'])
    mywriter.writerow(['float', 'float'])
    mywriter.writerow([','])
    
    for p in partStream:
        for note in p.flat.notesAndRests:
            if isinstance(note, music21.chord.Chord):
                pitch_val = note.root().midi
            elif isinstance(note, music21.note.Note):
                pitch_val = note.pitch.midi
            else: #rest
                continue
            row_info = [str(note.offset), str(convert_fifth(pitch_val))]
            mywriter.writerow(row_info)
```

```python
def convert_fifth(x):
    if x == -1:
        return 12
    else:
        return [0,7,2,9,4,11,6,1,8,3,10,5][x%12]
```
